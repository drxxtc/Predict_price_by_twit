```python
import pandas as pd
import nltk
import re
from nltk.stem import WordNetLemmatizer 
nltk.download('popular')
```

```python
data=pd.read_parquet('/content/drive/MyDrive/6195/2021-01')
```

```python
data
```

```python
data=data[['id','date','headline', 'timestamp', 'tags']]
```

```python
data['hashtag'] = data['headline'].apply(lambda x: re.findall(r"#(\w+)", x))
data['hashtag']
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for i in range(len(data['headline'])):
  data['headline'][i]=(lemmatizer.lemmatize( data['headline'][i])).lower()
```

```python
data['headline'].head()
```

—É–¥–∞–ª–µ–Ω–∏–µ –∑–Ω–∞–∫–æ–≤ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è –∏ —Ü–∏—Ñ—Ä

```python
for i in range(len(data)):
    data.headline[i] = re.sub(r'[^\w\s]', '', data.headline[i])
    data.headline[i] = re.sub(r'[\d]', '', data.headline[i])
data.head()
```

–£–¥–∞–ª–µ–Ω–∏–µ —Å—Ç–æ–ø-—Å–ª–æ–≤ –∏ —Ç–æ–∫–µ–Ω–∏–∑–∞—Ü–∏—è

```python
from nltk.corpus import stopwords
```

```python
for i in range(len(data)):
    temp_tokenize = []
    str_tokenize = ''
    temp_tokenize += nltk.word_tokenize(data.headline[i])
    for j in range(len(temp_tokenize)):
        if temp_tokenize[j] not in stopwords.words('english'):
            str_tokenize += temp_tokenize[j]
        if (j < len(temp_tokenize)):
            str_tokenize += ' '
    data.headline[i] = str_tokenize
data.head()
```
